```python
df["Item Fat Content"] = df["Item Fat Content"].replace({"LF": "Low Fat","low fat":"Low Fat"})
```

```python
df["Item Fat Content"] = df["Item Fat Content"].replace({"reg": "Regular","l})
```

```python
df['Item Fat Content'] = df['Item Fat Content'].fillna(df['Item Fat Content'].mode()[0])
```

```python
df['Item Identifier'] = df['Item Identifier'].fillna(df['Item Identifier'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Item Type"] = df["Item Type"].fillna(df["Item Type"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Outlet Establishment Year"] = df["Outlet Establishment Year"].astype('Int64')
```

```python
df
```

```python
df["Outlet Establishment Year"] = df["Outlet Establishment Year"].fillna(df["Outlet Establishment Year"].mode()[0])
```

```python
df["Outlet Identifier"] = df["Outlet Identifier"].fillna(df["Outlet Identifier"].mode()[0])
```

```python
df["Outlet Location Type"] = df["Outlet Location Type"].fillna(df["Outlet Location Type"].mode()[0])
```

```python
df["Outlet Size"] = df["Outlet Size"].fillna(df["Outlet Size"].mode()[0])
```

```python
df["Outlet Type"] = df["Outlet Type"].fillna(df["Outlet Type"].mode()[0])
```

```python
df["Item Visibility"] = df["Item Visibility"].fillna(df["Item Visibility"].median())
```

```python
df["Item Weight"] = df["Item Weight"].fillna(df["Item Weight"].mean())
```

```python
df["Sales"] = df["Sales"].fillna(df["Sales"].median())
```

```python
df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
```

```python
df["Outlet Age"] = 2026 - df["Outlet Establishment Year"]
```
